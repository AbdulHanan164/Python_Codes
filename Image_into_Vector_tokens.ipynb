```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Loading the image from the specific path

```python
#Mera naam
name = 'Abdul Hanan'

image_path = r"/content/Screenshot 2024-04-23 130335.png"
image = cv2.imread(image_path)
```

### Converting the image to grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Applying Canny edge detection to the grayscale image


```python
edges = cv2.Canny(gray_image, 100, 200)
print(edges)
```

### Finding contours in the edge-detected image

```python
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(cv2.RETR_EXTERNAL)
print(cv2.CHAIN_APPROX_SIMPLE)
print(contours)
```

### Converting contours to vectors (points)

```python
vectors = []
for contour in contours:
    for point in contour:
        vectors.append(point[0])

for i in range(len(vectors)):
    print(vectors[i], end='\t')
```

### Creating the visualization picture
